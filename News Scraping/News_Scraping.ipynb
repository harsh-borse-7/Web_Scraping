# News Data Scraping on The Wire

### Importing necessary libraries

```python
import requests
from bs4 import BeautifulSoup
```

### Extracting links from various pages

```python
section_list=['category/politics/all','category/economy/all','category/world/all','category/security/all','category/law/all','category/science/all','category/socity/all','category/culture/all']
```

### Gathering link at one place

```python
links_list=[] 
for link1 in section_list:
    url=f"https://thewire.in/{link1}"
    r= requests.get(url)
    htmlcontent = r.text
   
    soup = BeautifulSoup(htmlcontent, 'html.parser')
   

    links = soup.find_all('a')
   

    for link in links:
        href = link.get('href')
        url_link=f"https://thewire.in{href}"
        if url_link not in links_list:
            links_list.append(url_link)
links_list
```

### Scrapping the news and store in CSV file

```python
import csv
from bs4 import BeautifulSoup
import requests


with open('news_data1.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Link', 'Section', 'Heading', 'News']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()


    for lk in links_list:
        
        try:
            r = requests.get(lk)
            htmlcontent = r.text

            soup = BeautifulSoup(htmlcontent, 'lxml')

            section = soup.find('div', class_='tag')

            
            headings = soup.find('h1', class_='title').find('span')

            news = soup.find('div', class_="grey-text").find_all('p')

            str_news = []
            for text in news:
                if text:
                    str_news.append(text.text)

            news_paragraph = ' '.join(str_news)
            
          
            
            writer.writerow({'Link': lk, 'Section': section.text, 'Heading': headings.get_text(), 'News': news_paragraph})
            
        except Exception:
            pass  
```
